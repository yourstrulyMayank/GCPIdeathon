```python
model = YOLO(r"C:\Users\HP\Documents\GCP Reimagine\Trap-Camera Animal Detection - YOLO\runs\detect\train8\weights\best.pt")
```

```python
# Define path to video file
img = r'C:\Users\HP\Documents\GCP Reimagine\Trap-Camera Animal Detection\aug_data\test\images\Sun-Bear-Jamron-or-Bopha-Perth-Zoo.0.jpg'
res = model(img)
res_plotted = res[0].plot()
cv2.imshow("result", res_plotted)
```

```python

# Open the video file
video_path = r"C:\Users\HP\Desktop\Trap\Rhinoceros hornbill (Buceros rhinoceros).mp4"
cap = cv2.VideoCapture(video_path)


# Loop through the video frames
while cap.isOpened():
    # Read a frame from the video
    success, frame = cap.read()

    if success:
        # Run YOLOv8 inference on the frame
        results = model(frame)

        # Visualize the results on the frame
        annotated_frame = results[0].plot()

        # Display the annotated frame
        cv2.imshow("YOLOv8 Inference", annotated_frame)

        # Break the loop if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        # Break the loop if the end of the video is reached
        break

# Release the video capture object and close the display window
cap.release()
cv2.destroyAllWindows()
```
